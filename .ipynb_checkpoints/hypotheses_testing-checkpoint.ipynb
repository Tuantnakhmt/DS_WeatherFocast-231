## 2. Sau khi quan sát dữ liệu, ta có thể đặt ra vài giả thuyết

### 2.1. We may want to know how packaging affects ratings.

### Ta có thể thấy có 4 loại Style xuất hiện nhiều trong các review, và Pack (ramen gói) là dạng phổ biến nhất, tiếp đó là Bowl, Cup, cuối cùng là Tray. Để kiểm tra độ ảnh hưởng của các loại Style tới đánh giá của ramen, ta có thể kiểm tra xem chúng có cùng một phân phối hay không. Tuy nhiên, trước hết ta cần khảo sát xem dữ liệu hiện tại có tuân theo phân phối chuẩn, hay có thỏa mãn tính 'homogenity' hay chưa.

```python
# Bowl style vs Cup style
bowl_stars = df[df['Style'] == 'Bowl']['Stars']
cup_stars = df[df['Style'] == 'Cup']['Stars']
```

```python
### Để kiểm tra xem dữ liệu có tuân theo phân phối chuẩn hay không ###
### Ta có thể sử dụng Shapiro-Wilk test ###
### Null hypothesis: dữ liệu tuân theo phân phối chuẩn ###
### Nếu pValue < 0.05 ==> phản bác null hypothesis ###

### More information about the test could be found online :)) ###

bowl_dist = stats.shapiro(bowl_stars)
cup_dist = stats.shapiro(cup_stars)

print('pvalue for Bowl distribution: ', bowl_dist[1])
print('pvalue for Cup distribution: ', cup_dist[1])

### Ta có thể thấy đánh giá của Cup và Bowl không tuân theo phân phối chuẩn với độ tin cậy 95%. Điều này có thể được tháy rõ hơn ở đồ thị phía dưới ###
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].hist(x=bowl_stars, bins=15, color = 'g')
ax[0].title.set_text('Rate for Bowl Distribution')

ax[1].hist(x=cup_stars, bins=15, color = 'grey')
ax[1].title.set_text('Rate for Cup Distribution')

plt.show()

print('Here we clearly see that both the variables do not follow a normal distribution')
```

```python
### Quay trở lại ví dụ, ta sử dụng non-parametric mannwhitneyu test ###
### Test whether there is shift in location; Care about rank (not actual value) ###
### https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.mannwhitneyu.html ###
### http://www.compbio.dundee.ac.uk/user/mgierlinski/talks/pvalues2/p-values5.pdf ###

### Ta kiểm tra với 3 loại giả thuyết thay thế như sau ###
different = stats.mannwhitneyu(bowl_stars, cup_stars  , alternative='two-sided')
bowl_lt_cup = stats.mannwhitneyu(bowl_stars, cup_stars, alternative='less')
bowl_gt_cup = stats.mannwhitneyu(bowl_stars, cup_stars, alternative='greater')

print(different[1])
print(bowl_lt_cup[1])
print(bowl_gt_cup[1])

if different[1] < 0.05:
  print('Ratings are statistically not equal for Bowl and Cup at confidence level 0.05')

if bowl_lt_cup[1] < 0.05:
  print('Bowl statistically has less rate than Cup at confidence level 0.05')

if bowl_gt_cup[1] < 0.05:
  print('Bowl statistically has more rate than Cup at confidence level 0.05')
```

#### Bonus


```python
### Nếu dữ liệu không tuân theo phân phối chuẩn, ta có thể cân nhắc sử dụng non-parametric test (ở trường hợp này là Mann Whitnet U Test) ###
### Tuy nhiên giả sử ở bước trên giả thuyết null không bị bác bỏ, ta có thể chọn parametric test (2 sample t-test) ###
### Khi đó, ta cần kiểm tra nếu phân phối của 2 sample có cùng variance hay không (vì t-test cần có giả sử này) ###

### Để kiểm tra nếu các nhóm dữ liệu tuân theo các phân phối có cùng phương sai ###
### Ta có thể sử dụng levene test với H0: các phân phối dữ liệu có cùng phương sai ###
### Ngoài ra còn có các tests khác (Bartlett) ###
### Xem thêm tại https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.levene.html#scipy.stats.levene ###

lavene_test = stats.levene(bowl_stars, cup_stars)

print('pvalue for equal variance: ', lavene_test[1])
print('Variance of bowl rate', statistics.variance(bowl_stars))
print('Variance of cup rate', statistics.variance(cup_stars))
print('Var bowl / Var cup', statistics.variance(bowl_stars)/statistics.variance(cup_stars))

### Vì pvalue > 0.05, ta chưa đủ dữ kiện để bác bỏ H0 từ dữ liệu hiện có ###
### Nói cách khác, ta có thể giả sử các phân phối có cùng phương sai ###
### Một cách trực quan từ đồ thị bên trên, 2 phân phối gần như có cùng 'shape' ==> cùng phương sai ###
```

```python
### Một ví dụ tương tự với Pack và Bowl ###
pack_stars = df[df['Style'] == 'Pack']['Stars']

pack_dist = stats.shapiro(pack_stars)
print('pvalue for Pack Distribution: ', pack_dist[1])

different = stats.mannwhitneyu(bowl_stars, pack_stars  , alternative='two-sided')
bowl_lt_pack = stats.mannwhitneyu(bowl_stars, pack_stars, alternative='less')
bowl_gt_pack = stats.mannwhitneyu(bowl_stars, pack_stars, alternative='greater')

if different[1] < 0.05:
  print('Ratings are statistically not equal for the two samples at confidence level 0.05')
else:
  print('There is not sufficient evidence to conclude that the ratings of the two styles differ at confidence level 0.05.')

if bowl_lt_pack[1] < 0.05:
  print('Bowl statistically has less rate than Pack at confidence level 0.05')

if bowl_gt_pack[1] < 0.05:
  print('Bowl statistically has more rate than Pack at confidence level 0.05')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].hist(x=bowl_stars, bins=15, color = 'g')
ax[0].title.set_text('Rate for Bowl Distribution')

ax[1].hist(x=pack_stars, bins=15, color = 'grey')
ax[1].title.set_text('Rate for Pack Distribution')

plt.show()

print('Here we clearly see that both the variables do not follow a normal distribution, and have kind of same shapes.')
```

### Ta cũng quan sát được rằng hãng Nissin xuất hiện nhiều nhất trong dataset này. Vì thế ta có thể kiểm tra xem có sự tương quan nào giữa độ đa dạng của một hãng so với rate trung bình của nó.
### Ta có thể sử dụng Correlation tests như Pearson's correlation

```python
# Association between a brand's ratings and its number of varieties

tmp = df.groupby('Brand').agg({'Variety': 'count', 'Stars': np.mean})
print(tmp[tmp.Variety==tmp.Variety.max()])

correlation_factor, pvalue = stats.pearsonr(tmp["Variety"],tmp["Stars"])
print('correlation_factor = %.4f, pvalue = %.4f' % (correlation_factor, pvalue))

if pvalue < 0.05: print('There is statistically association between...')
else: print('There is not enough evidenve to support assiciation between....')

tmp.plot.scatter(x='Variety', y='Stars', c='DarkBlue')

### Ta cũng dễ dang quan sát vì Nissin ở đây như 1 outlier, nên ko tìm được độ tương quan nào giữa 2 thông tin. ###
```

### Một giả thuyết khác ta có thể kiểm tra là so sánh rate của các nước cùng có hãng Nissin. Ở đây 2 nước có nhiều review nhất được chọn. Ta sẽ tiếp tục sử dụng 2 samples test.

```python
# Tìm 2 nước có nhiều review nhất và cùng có hãng Nissin

nissin = df[df.Brand == 'Nissin']
list_countries = nissin.Country.value_counts()[:2].index.to_list()
print(list_countries)
first_country_rates = nissin[nissin.Country == list_countries[0]].Stars
second_country_rates = nissin[nissin.Country == list_countries[1]].Stars

# Creating plot
fig = plt.figure(figsize =(10, 7))
bp = plt.boxplot([first_country_rates, second_country_rates])
plt.show()

# Check for normality
first_country_dist = stats.shapiro(first_country_rates)
second_country_dist = stats.shapiro(second_country_rates)
print('pvalue for Japan Distribution: ', first_country_dist[1])
print('pvalue for USA Distribution: ', second_country_dist[1])

# Use non-parametric test
different = stats.mannwhitneyu(first_country_rates, second_country_rates  , alternative='two-sided')
j_lt_u = stats.mannwhitneyu(first_country_rates, second_country_rates, alternative='less')
j_gt_u = stats.mannwhitneyu(first_country_rates, second_country_rates, alternative='greater')

if different[1] < 0.05:
  print('We have statistically significant evidence at α =0.05 to show that the rating populations of two countries are not equal.')
else:
  print('There is no significant difference between rating distributions of the two countries.')

if j_lt_u[1] < 0.05:
  print('Japan statistically has less rate than the USA at confidence level 0.05')

if j_gt_u[1] < 0.05:
  print('Japan statisticcally has more rate than the USA at confidence level 0.05')
```

### 2.4.Ta sẽ thử khai thác thông tin từ trường Variety. Trường này mô tả về các loại ramen, và có thể được tận dụng để khai thác thông tin về nguyên liệu/hương vị.

### Để lấy được các từ liên quan tới nguyên liệu, ở đây đơn giản ta sẽ tách câu thành các từ, sau đó chọn 2 từ 'hợp lý' có tần số cao nhất.


```python
from sklearn.feature_extraction.text import CountVectorizer
def get_top_n_gram(corpus, ngram, stop_words=None):
    """
    Generates a bag of words as list of tuples for a corpora of words
    :param corpus: list of words to apply bag of words
    :param ngram: int the ngram to be returned
    :param stop_words: list of stop words to remove from bag of words
    :return: Bag of words as a list of tuples sorted in descending order of frequency
    """
    vec = CountVectorizer(stop_words=stop_words, ngram_range=(ngram, ngram))
    bag_of_words = vec.fit_transform(corpus)
    counts_of_words = np.array(bag_of_words.sum(axis=0))[0]          # calcaulate the number of occurrances of each word
    words_freq = [(word, counts_of_words[idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)
    return words_freq

stop_words = []                      # I use standard stop_words of NLTK
with open('stop_words.txt') as f:
    stop_words = f.readlines()
    stop_words = [word[:-1] for word in stop_words]

words_freq = get_top_n_gram(df.Variety, 1, stop_words)

print(words_freq[:20])

### Nhìn nhanh ta thấy kết quả khá hợp lý khi các từ như 'noodles', 'ramen', 'flavour' xuất hiện nhiều nhất. ###
### Tuy nhiên để phản ánh đúng về nguyên liệu, ta chọn 'chicken' và 'beef' ###
```

```python
### First, convert those desctiptions to lower case ###
tmp = df.copy()
tmp.Variety = tmp.Variety.str.lower()
tmp.head()
```

```python
### Tìm các review mà nguyên liệu có cả 'chicken' và 'beef'. Có thể những review này sẽ cần loại bỏ.###
tmp[tmp.Variety.str.contains('chicken') & tmp.Variety.str.contains('beef')]
```

```python
### Chọn ra rate tương ứng của review chứa 'chicken' và 'beef' trong Variety ###

# chicken vs beef
chicken_rate = tmp[tmp.Variety.str.contains('chicken')].Stars
beef_rate = tmp[tmp.Variety.str.contains('beef')].Stars

chicken_dist = stats.shapiro(chicken_rate)
beef_dist = stats.shapiro(beef_rate)
print('pvalue for chicken rate Distribution: ', chicken_dist[1])
print('pvalue for beef rate Distribution: ', beef_dist[1])

different = stats.mannwhitneyu(chicken_rate, beef_rate  , alternative='two-sided')
c_lt_b = stats.mannwhitneyu(chicken_rate, beef_rate, alternative='less')
c_gt_b = stats.mannwhitneyu(chicken_rate, beef_rate, alternative='greater')

if different[1] < 0.05:
  print('The 2 distributions are statisticcally different at confident level 0.05')
else:
    print("There is not sufficient evidence to conclude that the ratings of 'chicken flavour' and 'beef flavour' differ.")
if c_lt_b[1] < 0.05:
  print('Chicken statistically has less rate than beef at confident leve 0.05')

if c_gt_b[1] < 0.05:
  print('Chicken statictically has more rate than the beef at confident level 0.05')
```
