```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('weather_city.csv')
data
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['pres','slp','vis'], inplace=True)
```

```python
data.isnull().sum()
```

```python
data = data.drop(['Unnamed: 0.3','Unnamed: 0.2','Unnamed: 0.1','Unnamed: 0', 'timestamp_utc','revision_status','pod','ts'], axis = 1)
```

```python
data.info()
```

```python
#Xử lý thời gian:
import datetime as dt
from datetime import timedelta
data["Date"]=pd.to_datetime(data['timestamp_local'])
```

## EDA

### Phân tích về độ biến động của từng đặc trưng

```python
#Tạo 1 dataframe cho 1 thành phố cụ thể là NewYork
weather_data = pd.DataFrame(data)
# Creating a new DataFrame for 'Sydney' city
newyork_weather = data[data['name'] == "New York City"]
```

```python
newyork_weather
```

```python
features = ['app_temp', 'clouds', 'dewpt', 'precip', 'rh', 'temp']

fig, axs = plt.subplots(len(features), figsize=(10, 8), sharex=True)
fig.suptitle('Biến động giá trị các đặc trưng app_temp, clouds, dewpt, precip, rh, temp theo thời gian')

for i, feature in enumerate(features):
    axs[i].plot(newyork_weather['Date'], newyork_weather[feature])
    axs[i].set_ylabel(feature)

plt.xlabel('Date')
plt.show()
```

```python
features = ['azimuth','dhi','elev_angle','ghi','solar_rad','uv']

fig, axs = plt.subplots(len(features), figsize=(10, 8), sharex=True)
fig.suptitle('Biến động giá trị của azimuth, dhi, elev_angle, ghi, solar_rad, uv theo thời gian tại NYC')

for i, feature in enumerate(features):
    axs[i].plot(newyork_weather['Date'], newyork_weather[feature])
    axs[i].set_ylabel(feature)

plt.xlabel('Date')
plt.show()
```

```python
features = ['wind_dir','wind_gust_spd','wind_spd']

fig, axs = plt.subplots(len(features), figsize=(10, 8), sharex=True)
fig.suptitle('Biến động các giá trị về wind_dir, wind_gust_spd, wind_spd theo thời gian ở NYC')

for i, feature in enumerate(features):
    axs[i].plot(newyork_weather['Date'], newyork_weather[feature])
    axs[i].set_ylabel(feature)

plt.xlabel('Date')
plt.show()
```

```python
features = ['pres','slp']

fig, axs = plt.subplots(len(features), figsize=(10, 8), sharex=True)
fig.suptitle('Biến động slp và pres theo thời gian')

for i, feature in enumerate(features):
    axs[i].plot(newyork_weather['Date'], newyork_weather[feature])
    axs[i].set_ylabel(feature)

plt.xlabel('Date')
plt.show()
```

Quan sát (tại New York): 
* Những đặc trưng phụ thuộc nặng nề theo thời gian trong ngày, diễn ra theo chu kỳ trong ngày thì biểu đồ dầy, mượt và đối xứng. Những đặc trưng này không bị phụ thuộc vào các đặc điểm khí tượng khác.
* Những đặc trưng khí tượng như app_temp, temp, dewpt, percip có xu hướng đối xứng. Đây là những đặc trưng diễn ra theo chu kỳ trong năm, do đó ít trơn mịn hơn.
* Các đặc trưng ảnh hưởng theo năm như wind_dir, wind_gust_spd, slp, pres có biên động lớn hơn ở các năm với nhau. Tuy vậy, ở các thành phố khác thì đặc trưng slp, pres lại ít biến động hơn.

```python
#correlation
data_corr=newyork_weather.drop(['timestamp_local','weather_description','name','lat','lon','weather_code'], axis = 1)

relation = data_corr.corr()
plt.figure(figsize=(15,10))
sns.heatmap(data=relation,annot=True,cmap='coolwarm')
plt.title("Tương quan đặc trưng tại NYC")
# # save the figure.
# plt.savefig('plot1.png', dpi=300, bbox_inches='tight')
plt.show()
```

Quan sát tại NYC: 
* Các đặc trưng địa lý diễn ra theo ngày như quan sát trước đó:'azimuth','dhi','elev_angle','ghi','solar_rad','uv' thường có ít mối quan hệ đến các đặc trưng khác. Lưu ý nhóm chỉ số 'dhi','elev_angle','ghi','solar_rad','uv' liên quan với nhau bởi khái niệm lượng bức xạ nên phụ thuộc rõ rệt (tỉ lệ thuận)
* Các cặp đặc trưng tỉ lệ nghịch với nhau lớn <-0,30: 
    -  clouds & solar_rad, clouds & win_dir
    -  dewpt & pres, dewpt & slp 
    -  pres & temp, slp & temp
    -  rh & solar_rad, rh & uv
* Các đặc trưng tỉ lệ thuận với nhau > 0,5:
    -  app_temp &dewpt, app_temp &temp
    -  dewpt &temp
    -  dhi & elev_angle, dhi &ghi, dhi & solar_rad, dhi & uv,...
    -  rh & dewpt

### Nhãn và các đặc trưng (xét tại NYC)

```python
Weather= newyork_weather['weather_description'].value_counts().reset_index()
Weather.columns=['Weather','Count']
Weather
```

Chúng ta có tận 17 nhãn thời tiết, số lượng mẫu cho từng nhãn rất mất cân bằng (lớp thiểu số chiếm <1%). Thời tiết ở nơi này đa phần liên quan về mây.  

```python
sns.set(rc={'figure.figsize':(8,4)})
plt.xticks(rotation=90)
sns.lineplot(x=Weather['Weather'],y=Weather['Count'],data=Weather)
plt.show()
```

```python
#Weather vs Temp 
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title('Weather and Temperature')
sns.barplot(x=data['weather_description'],y=data['temp'])
```

Đối với đặc trưng temp, độ lệch chuẩn trên toàn bộ mẫu của từng nhãn không lớn, riêng với nhãn Heavy snow có độ lệch chuẩn lớn nhất, chứng tỏ các giá trị của temp sai khác nhiều với từng mẫu nhãn này.

```python
#Weather vs cloud
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title('Weather and cloud')
sns.barplot(x=data['weather_description'],y=data['clouds'])
```

Đặc trưng clouds cũng tương tự, nhưng đặc biệt với xu hướng lượng mây tăng dần thì số lượng mẫu cho từng nhãn liên quan về lượng mây cũng tăng.

```python
#Weather vs DHI
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title('Weather and DHI')
sns.barplot(x=data['weather_description'],y=data['dhi'])
```

Đối với đặc trưng DHI, độ lệch chuẩn trên các nhãn dữ liệu khá lớn, đặc biệt lớn với Haze, và các nhãn liên quan đến rain và snow

```python
#Weather vs pressure
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title('Weather and pressure')
sns.lineplot(x=data['weather_description'],y=data['pres'])
```

```python
#Weather vs Relative humid
plt.figure(figsize=(12,6))
plt.xticks(rotation=90)
plt.title('Weather and Relative humid')
sns.lineplot(x=data['weather_description'],y=data['rh'])
```

Xu hướng: thời tiết từ mây đến mưa đến sương mù thì độ ẩm tăng theo.

Tóm lại:
* Các đặc trưng ảnh hưởng theo ngày làm cho mỗi nhãn lớp có độ lệch chuẩn lớn, khó để nhận biết được sau này
* Các đặc trưng quan trọng như clouds, rh, temp,... thể hiện cụ thể những nhãn lớp cụ thể hơn.

```python
# Pivot the DataFrame for easy plotting
pivot_data = NYC_beijing_weather.pivot_table(index='Date', columns='name', values=['pres', 'rh'])

# Plotting
plt.figure(figsize=(10, 6))

# Pressure
plt.subplot(2, 1, 1)
pivot_data['pres'].plot(ax=plt.gca())
plt.title('Pressure Over Time')
plt.ylabel('Pressure')

# Sea-Level Pressure
plt.subplot(2, 1, 2)
pivot_data['rh'].plot(ax=plt.gca())
plt.title('rh Over Time')
plt.ylabel('rh')

plt.tight_layout()
plt.show()
```
